```python
import requests
from bs4 import BeautifulSoup as bs
import re
import pandas as pd
from time import sleep
```

```python
dstore = []
for i in range(1, 9):
    print(i)
    res = requests.get(f'https://election.dhakatribune.com/get-seats/{i}')
    if res.status_code == 200:
        print('200')
        data = res.json()
        for seat in data:
            dstore.append(seat)

dfd = pd.DataFrame(dstore)
dfd.head()
```

```python
dfd.shape
```

```python
dfd.to_csv('data/dt_seats.csv', index=False)
```

```python
rstore = []
count = 5
for i, row in dfd.iterrows():
    print(row.seat_name)
    res = requests.get(f'https://election.dhakatribune.com/get-candidate/{row.seat_no}')
    print(res.status_code)
    if res.status_code == 200:
        data = res.json()
        for candidate in data:
            rstore.append(candidate)
    count = count - 1
    if count == 0:
        print('sleeping')
        sleep(2)
        count = 5
```

```python
dfr = pd.DataFrame(rstore)
dfr.head()
```

```python
party_mapping = {
    '38': 'BNP',
    '36': 'Jamaat',
    '66': 'AB Party',
    '37': 'Islami Andolan Bangladesh',
    '5': 'JaPa',
    '47': 'CPB',
    '90': 'Ganosamhati Andolon (Saki)',
    '63': 'NCP',
    '50': 'Bangladesh Khelafat Majlish',
    '115': 'Gano Forum',
    '99': 'Jamiat Ulema-e-Islam Bangladesh',
    '45': 'Khelafat Majlish',
    '65': 'BJP'
}

def map_party(x):
    x = str(x)
    if x in party_mapping.keys():
        return party_mapping[x]
    return None

dfr['party'] = dfr.party_id.apply(map_party)
dfr.head()
```

```python
dfr = dfr.dropna(subset=['party'])
dfr.head()
```

```python
df = pd.merge(dfr, dfd, how='left', left_on='seat_id', right_on='seat_no')
df = df[['seat_name', 'party', 'name', 'vote', 'center', 'division_id', 'total_center']]
df.head()
```

```python
seat_mapping = {
    'Chapainawabganj': 'Chapai Nawabganj',
    'Jhalakathi': 'Jhalokati'
}

def map_seat(x):
    name, number = x.split('-')
    if name in seat_mapping.keys():
        return seat_mapping[name] + '-' + number
    return x

df['seat_name'] = df.seat_name.apply(map_seat)
df.head()
```

```python
df['time'] = pd.to_datetime('now', utc=True)
df['source'] = 'dt'
df.head()
```

```python
df.to_csv('data/votes_raw.csv', index=False)
```
