```python
import pandas as pd
import json
```

```python
with open('pa.json', 'r', encoding='utf8') as f:
    data = json.load(f)
    seats = data['resultData']['win_nearest_candidates']
    print(len(seats))
```

```python
store = []
for seat_no, seat in seats.items():
    if 'nearest' in seat:
        nearest = seat['nearest']
        store.append((
            nearest['area_name'],
            nearest['candidates_name'],
            nearest['party'],
            nearest['vote_cast'],
            nearest['jote']
        ))
    if 'win' in seat:
        win = seat['win']
        store.append((
            win['area_name'],
            win['candidates_name'],
            win['party'],
            win['vote_cast'],
            win['jote']
        ))
    
print(len(store))
```

```python
df = pd.DataFrame(store, columns=['seat', 'name_bn', 'party_bn', 'vote', 'jote'])
df.head()
```

```python
df.to_csv('pa.csv', index=False)
```
